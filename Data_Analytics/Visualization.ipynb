```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [88,78,78,23,34,23,66]
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(y)
plt.show()
```

```python
fig, ax=plt.subplots(nrows=2, ncols=1)
plt.show()
```

```python
t = np.linspace(0, 4*np.pi,60 )
y = np.sin(t)
x = np.cos(t)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,2))
ax[0].plot(t,x, 'r')
ax[1].bar(t,y)
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,56]
plt.plot(x,y, linewidth = 2, color ='purple', marker='o', markersize = '15', markerfacecolor= 'white' , markeredgecolor='red', linestyle="--")
plt.show()
```

```python
plt.plot(x,y, linewidth = 2, color ='purple', marker='o', markersize = '15', markerfacecolor= 'white' , markeredgecolor='red', linestyle="--")
plt.title('A Random Graph', fontdict={'fontsize':15, 'fontweight':'bold', 'color':'green'})
plt.xlabel("Values of X ----->", fontdict={'color':'red'})
plt.ylabel("Values of Y ----->", fontdict={'color':'red'})
plt.text(800,60,"This is a Text")
plt.annotate("Peak Value",xy=(500,65), xytext=(600,50), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```
