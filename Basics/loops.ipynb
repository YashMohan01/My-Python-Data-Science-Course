## break and continue

```python
x = [1, 2, 0, 3, 4, 0, 0, 5]
```

```python
for i in x:
    if i == 0:
        break
    print(i)
```

```python
for i in x:
    if i == 0:
        continue
    print(i)
```

```python
for i in x:
    if i != 0:
     print(i)
```

# for-else

```python
num = 11
for i in range(2  , num):
    if num % i == 0:
        print('not prime')
        break
else:
    print('prime')
```

```python
for num in range(1, 100):
    if num == 1:
        continue
    for i in range(2 , num):
        if num % i == 0:
            break
    else:
        print(num, end='  ') 
```

```python
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print('done')
```

# while loop

```python
x = 10
while x > 0:
     x -= 2
     print(x)   
```

```python
total = 0
while True:
    num = int(input('Enter a number:'))
    if num == 0:
        break
    total += num
print(f'Total is {total}')
```

```python
books=[]
while True:
    name = input('Enter book name:')
    if name:
        books.append(name)
    else:
        break
print("These are your books")
print(books)
```
